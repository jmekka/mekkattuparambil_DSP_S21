##### a) Explain how k-fold cross-validation is implemented.

The k-fold cross validation approach divides the original data set into k groups about the same size. One group is treated as a test data set and the other k-1 groups are used as a training data set. The model will be fit on the training set then computed on the test set, so one can estimate how well the model fits on new data.

##### b) What are the advantages and disadvantages of k-fold crossvalidation relative to:
##### i. The validation set approach?
##### ii. LOOCV?

b) i. Test error rates of a k-fold CV are not highly variable as with the validation set approach, since the validation set approach depends on which observation group is used as the training and test data set. Also it tends to overestimate the test error rate of the model fit on the whole data set. 

ii. The k-fold CV has computational advantages to LOOCV, since the latter can be time consuming depending on the sample size. Another advantage lies in bias-variance trade-off:k-fold CV using k=5 or k=10 don't suffer from high bias or high variance.
