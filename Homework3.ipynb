The dataset "health_m" includs 865 men and the data "health_w" includes 921 women.

The datasets are not tidy, because the same observational unit are saved in separate tables. To make the datasets tidy, one need to put the two datasets into on single table.

```r
health_mw <- bind_rows(health_m, health_w, .id='gender')
health_mw
```

```r
ggplot(health_mw,aes(bmi)) +
 geom_density()
```

The two peaks in the density plot displays where the values of the variable 'bmi' are concentrated. 

```r
t.test(bmi ~ gender, data = health_mw)
```

```r
ggplot(health_mw,aes(bmi, color=gender)) +
 geom_density()
```

The t-test does not help to describe the shape of the density plot, since it visualises the distribution of the data. The t-test however tests for the differences in the means. The t-test shows that men and women differ in their bmi. The p-value (0.001238) is lower than the significance level of 0.05. 

```r
hist(health_mw$steps)
```

It is unusual that there is a data point with the value "-5'000".

```r
health <- filter(health_mw, steps>=0)
hist(health$steps)
```

```r
cor.test(health$steps, health$bmi)
```

There is a statistically significant correlation between the variables 'steps' and 'bmi', p(2.2e-16)<0.05. 

```r
health_m <-filter(health, gender==1)
health_w <-filter(health, gender==2)
cor.test(health_m$steps, health_m$bmi)
cor.test(health_w$steps, health_w$bmi)
```

Yes, both groups have different correlation between the variables 'steps' and 'bmi'.
